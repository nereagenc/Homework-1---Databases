```python
file_name = "data.txt"

try:
    f = open(file_name, "w")
    f.write("Record 1: User=Nerea\n")
    f.write("Record 2: Balance=1000\n")

    raise Exception("Simulated crash during file write")

    f.write("Record 3: Status=Complete\n")
    f.close

except Exception as e:
    print ("Program crashed:", e)
```

For this part, I wrote a file (data.txt) one line at a time, like it was saving records. After writing the first two lines, I forced the program to "crash" by raising an exception before it could finish writing the last line. When it crashes, the file still exists, but it's missing part of the data, so it's not in the state the program intended. The problem with plain files is there's no built-in "all-or-nothing" protection--writing is not automatically treated like a safe transaction. After a failure, you can't easily tell if the file is complete or only partially written just by looking at it. To be sure, you'd need extra steps like writing to a temporary file and only replacing the real file after everything suceeds.



**Key question answer:** You don't really know the file is correct after a crash unless you add your own safety checks.
