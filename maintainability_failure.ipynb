```python
file_name = "users_v2.txt"

f = open(file_name, "r")
data = {}

for line in f:
    key, value = line.strip().split("=")
    data[key] = value

f.close()
print("Parsed data", data)
```

For Part 3, I created two versions of the same data file to show how file formats become hard to maintain over time. In the first version, each line uses the format ke=value, so my parser reads each line and splits on "=" to get the key and value. In the second version, the structure changed, the separator changed to ":" and new fields were added, but the parsing code did not change. Because of that, the program crashes with a ValueError when it tries to split a line that no longer contains "=". This becomes error-prone because a small, "reasonable" format change breaks older code immediatley, adn there is no built-in schema or version label in a plain file to warn you that the format is different. Future developers get stuck either hunting down why the parser broke or adding more and more special-case rules for each format, which makes the code harder to understand, test, and modify safely.

**Key question answer:** This design fails future developers beacuse the file format can change silently and every reader of the file has to guess the format and update their code manually.
